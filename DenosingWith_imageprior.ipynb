```python
import os
import numpy as np
import tensorflow as tf
from skimage import io
from skimage.transform import resize
from tqdm.keras import TqdmCallback
import matplotlib.pyplot as plt

# Function to load and preprocess the image
def load_and_preprocess_image(img_path, target_size=(256, 256)):
    """Load and preprocess the image (resize, normalize)."""
    img = io.imread(img_path).astype(np.float32) / 255.  # Normalize to range [0, 1]
    img_resized = resize(img, target_size, mode='reflect', anti_aliasing=True)  # Resize image
    return img_resized

# Function to add noise to the image
def get_noisy_img(img, sig=30):
    """Add white noise to the image."""
    sigma = sig / 255.
    noise = np.random.normal(scale=sigma, size=img.shape)
    img_noisy = np.clip(img + noise, 0, 1).astype(np.float32)
    return img_noisy

# Function to calculate PSNR (Peak Signal to Noise Ratio)

def psnr_to_gt(x_true, x, maxv=1.):
    """Calculate PSNR between true and generated image."""
    mse = tf.reduce_mean(tf.keras.losses.MeanSquaredError()(x, x_true))  # Update here
    psnr_ = 10. * tf.math.log(maxv** 2 / mse) / tf.math.log(10.)
    return psnr_

# Deep Image Prior workflow for denoising
def dip_workflow(x0, x_true, f, f_input_shape, z_std=0.1, loss_mask=None, num_iters=1, init_lr=0.01, save_filepath=None):
    """Deep Image prior denoising workflow"""
    # Sample input z
    shape = (1,) + f_input_shape
    z = tf.constant(np.random.uniform(size=shape).astype(np.float32) * z_std, name='net_input')

    # Training Loss function
    def loss_fn(x_true, x):
        del x_true
        nonlocal x0, loss_mask
        if loss_mask is None:
            return tf.keras.losses.MSE(x, x0)
        else:
            return tf.keras.losses.MSE(x * loss_mask, x0 * loss_mask)

    # Optimization
    opt = tf.keras.optimizers.Adam(learning_rate=init_lr)
    f.compile(optimizer=opt, loss=loss_fn, metrics=[psnr_to_gt])

    # Custom callback to print epoch number during training
    class EpochPrintCallback(tf.keras.callbacks.Callback):
        def on_epoch_end(self, epoch, logs=None):
            print(f"Epoch {epoch + 1}/{num_iters} completed")

    # Training the model
    history = f.fit(z, x_true[None, ...], epochs=num_iters, steps_per_epoch=1, verbose=0, callbacks=[TqdmCallback(verbose=1), EpochPrintCallback()])

    # Display the results
    x = f.predict(z)[0]
    fig = plt.figure(figsize=(15, 5))  # Wider figure for 3 images in a row
    axes = [fig.add_subplot(1, 3, i + 1) for i in range(3)]  # Create subplots in a row

    # Plot the images
    axes[0].imshow(x0); axes[0].set_title('Input noisy image'); axes[0].axis('off')
    axes[1].imshow(x_true); axes[1].set_title('Ground-truth image'); axes[1].axis('off')
    axes[2].imshow(x); axes[2].set_title(f'Last output (PSNR = {psnr_to_gt(x_true, x):.2f})'); axes[2].axis('off')

    plt.show()
    return x

class GaussianNoiseWithDecay(tf.keras.layers.GaussianNoise):
    def __init__(self, stddev, decayrate=0.99999, decaysteps=1, **kwargs):
        super(GaussianNoiseWithDecay, self).__init__(stddev, **kwargs)
        self.num_calls = 0
        self.decayrate = decayrate
        self.decaysteps = decaysteps

    def call(self, inputs, training=None):
        def noised():
            self.num_calls += 1
            stddev = self.stddev * self.decayrate ** (self.num_calls // self.decaysteps)
            return inputs + tf.keras.backend.random_normal(shape=tf.shape(inputs), mean=0., stddev=stddev, dtype=inputs.dtype)

        # Use 'training' directly instead of 'in_train_phase'
        return tf.where(training, noised(), inputs)

# Deep Image Prior model definition
def deep_image_prior(input_shape, noise_reg=None, layers=(128, 128, 128, 128, 128), kernel_size_down=3, kernel_size_up=3, skip=(0, 4, 4, 4, 4)):
    def norm_and_active(x):
        x = tf.keras.layers.BatchNormalization()(x)
        x = tf.keras.layers.LeakyReLU()(x)
        return x
    
    model = tf.keras.models.Sequential(name="Deep Image Prior")
    inputs = tf.keras.Input(shape=input_shape)
    
    # Initialize x with the input tensor
    x = inputs  # Initialize x with inputs
    
    # Add noise regularization if specified
    if noise_reg is not None:
        x = GaussianNoiseWithDecay(**noise_reg)(x)
    
    # Downsampling layers
    down_layers = []
    for i, (num_filters, do_skip) in enumerate(zip(layers, skip)):
        if do_skip > 0:
            down_layers.append(norm_and_active(tf.keras.layers.Conv2D(filters=do_skip, kernel_size=1, strides=1, name=f"conv_skip_depth_{i}")(x)))
        for j, strides in enumerate([2, 1]):
            x = norm_and_active(tf.keras.layers.Conv2D(num_filters, kernel_size_down, strides=strides, padding='same', name=f"conv_down_{j + 1}_depth_{i}")(x))
        
    # Upsampling layers
    for i, (num_filters, do_skip) in enumerate(zip(layers[::-1], skip[::-1])):
        x = tf.keras.layers.UpSampling2D(interpolation='bilinear', name=f"upsample_depth_{i}")(x)
        if do_skip:
            x = tf.keras.layers.Concatenate(axis=-1)([x, down_layers.pop()])
        for j, kernel_size in enumerate([kernel_size_up, 1]):
            x = norm_and_active(tf.keras.layers.Conv2D(num_filters, kernel_size, strides=1, padding='same', name=f"conv_up_{j + 1}_depth_{i}")(x))
    
    # Final output layer
    x = tf.keras.layers.Conv2D(filters=3, kernel_size=1, strides=1, name="conv_out")(x)
    x = tf.keras.layers.Activation('sigmoid')(x)
    
    return tf.keras.Model(inputs=inputs, outputs=x, name="deep_image_prior")


# Display the model architecture
def display_dip_model(input_shape=(256, 256, 3)):
    model = deep_image_prior(input_shape)
    model.build(input_shape)
    print(model.summary())
    

# Load the image and apply preprocessing
img_path = "/kaggle/input/testseg/00054138.png"  # Updated image path
img_true = load_and_preprocess_image(img_path)  # Preprocessed image
img_noisy = get_noisy_img(img_true)  # Add noise to the image

# Define model input shape and noise regularization parameters
input_shape = img_true.shape
noise_reg = {'stddev': 1./30., 'decayrate': 1.0, 'decaysteps': 100}

# Initialize and display the Deep Image Prior model
display_dip_model(input_shape)

# Create and run the Deep Image Prior workflow
model = deep_image_prior(input_shape, noise_reg=noise_reg)
x = dip_workflow(img_noisy, img_true, model, input_shape, num_iters=5000, save_filepath='best_dip_denoising.hdf5')
```
